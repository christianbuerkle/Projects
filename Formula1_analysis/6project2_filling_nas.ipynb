```python
import pandas as pd

```

```python
import fancyimpute
```

```python
from fancyimpute import KNN
```

```python
df = pd.read_csv('project2_season_level_clean.csv')
```

```python
del df['Unnamed: 0']
```

```python
df.isna().sum()
```

```python
df[df['points2'].isna()]
```

```python
df['championship_points'] = df['championship_points'].fillna(0)
```

```python
df['points2'] = df['points2'].fillna(0)
```

```python
df = df[df['points2'].notnull()]
```

```python
df2 = fancyimpute.KNN(k=5).fit_transform(df)
```

```python
df3 = pd.DataFrame(df2, columns = df.columns)
```

```python
df3.to_csv('project2_KNN_v1.csv')
```
